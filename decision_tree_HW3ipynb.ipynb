# **Cost-complexity Pruning**

Cost-complexity pruning is a technique used to improve the performance of decision tree models by pruning the tree to avoid overfitting. It does this by finding a subset of the tree that optimizes the trade-off between the tree's complexity (size) and its performance on the training data.

First demonstrate the cross validation technique to find out what number to use for ccp_alpha

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Load the iris data
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=42)

# Define a range of possible values for ccp_alpha
param_grid = {'ccp_alpha': [0.0001, 0.001, 0.01, 0.1, 1.0]}

# Create a DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=42)

# Set up GridSearchCV
grid_search = GridSearchCV(clf, param_grid, cv=5)

# Fit the GridSearchCV object to the data
grid_search.fit(X_train, y_train)

# The best ccp_alpha found
best_ccp_alpha = grid_search.best_params_['ccp_alpha']
print(f"The best value for ccp_alpha is: {best_ccp_alpha}")
```

```python
def train_model_with_ccp(iris, ccp_alpha_value=0.01):
    """
    Train decision tree classifier with cost-complexity pruning

    :param iris:    iris data set instance
    :param ccp_alpha_value:  The alpha value for cost-complexity pruning
    :return:        classifier instance
    """
    clf = tree.DecisionTreeClassifier(ccp_alpha=ccp_alpha_value)
    clf = clf.fit(iris.data, iris.target)
    return clf
```

# **Cross validation**

To check if your model improved with cost-complexity pruning (CCP), you should compare the performance of the pruned model against the original model on a held-out test set or use cross-validation

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier

# Load your data
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)

# Train the unpruned model
unpruned_clf = DecisionTreeClassifier(random_state=42)
unpruned_clf.fit(X_train, y_train)

# Train the pruned model
pruned_clf = DecisionTreeClassifier(ccp_alpha=0.01, random_state=42)  # adjust ccp_alpha based on cross-validation
pruned_clf.fit(X_train, y_train)

# Evaluate both models
unpruned_pred = unpruned_clf.predict(X_test)
pruned_pred = pruned_clf.predict(X_test)

# Compare their performance
unpruned_accuracy = accuracy_score(y_test, unpruned_pred)
pruned_accuracy = accuracy_score(y_test, pruned_pred)

print(f"Unpruned Model Accuracy: {unpruned_accuracy}")
print(f"Pruned Model Accuracy: {pruned_accuracy}")

# Determine if the pruned model is an improvement
if pruned_accuracy >= unpruned_accuracy:
    print("The model improved with CCP.")
else:
    print("The model did not improve with CCP.")
```

# **Using a Bagging Classifier**
Bagging (Bootstrap Aggregating) involves training the same algorithm many times using different subsets of the training dataset. In Python, you can use the BaggingClassifier class from sklearn.ensemble.

```python
import pydotplus
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from IPython.display import Image, display

def load_data_set():
    """
    Loads the iris data set

    :return: data set instance
    """
    iris = load_iris()
    return iris

def train_decision_tree(iris):
    """
    Train decision tree classifier

    :param iris: iris data set instance
    :return: classifier instance
    """
    clf = tree.DecisionTreeClassifier()
    clf = clf.fit(iris.data, iris.target)
    return clf

def train_bagging_model(iris):
    """
    Train a bagging ensemble of decision tree classifiers

    :param iris: iris data set instance
    :return: trained bagging classifier instance
    """
    base_estimator = tree.DecisionTreeClassifier()
    bagging_clf = BaggingClassifier(estimator=base_estimator, n_estimators=100, random_state=42)
    bagging_clf.fit(iris.data, iris.target)

    return bagging_clf

def display_image(clf, iris):
    """
    Displays the decision tree image
    :param clf: classifier instance
    :param iris: iris data set instance
    """
    dot_data = tree.export_graphviz(clf, out_file=None,
                                    feature_names=iris.feature_names,
                                    class_names=iris.target_names,
                                    filled=True, rounded=True)

    graph = pydotplus.graph_from_dot_data(dot_data)
    display(Image(data=graph.create_png()))

if __name__ == '__main__':
    iris_data = load_data_set()

    # Train the decision tree classifier
    decision_tree_classifier = train_decision_tree(iris_data)
    display_image(clf=decision_tree_classifier, iris=iris_data)

    # Train the bagging classifier
    bagging_classifier = train_bagging_model(iris_data)

    # Split the data
    X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, test_size=0.3, random_state=42)

    # Predict and evaluate the bagging classifier
    y_pred = bagging_classifier.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Bagging Classifier Accuracy: {accuracy}")
    print(classification_report(y_test, y_pred))
```

**Compare both models and their performances**

```python
from sklearn.metrics import accuracy_score, classification_report

# Assuming you have already split your data into training and testing sets (X_train, X_test, y_train, y_test)

# Train the decision tree classifier
decision_tree_classifier = train_decision_tree(iris_data)
decision_tree_classifier.fit(X_train, y_train)

# Train the bagging classifier
bagging_classifier = train_bagging_model(iris_data)
bagging_classifier.fit(X_train, y_train)

# Evaluate the decision tree
dt_predictions = decision_tree_classifier.predict(X_test)
dt_accuracy = accuracy_score(y_test, dt_predictions)
print(f"Decision Tree Accuracy: {dt_accuracy}")
print(classification_report(y_test, dt_predictions))

# Evaluate the bagging classifier
bagging_predictions = bagging_classifier.predict(X_test)
bagging_accuracy = accuracy_score(y_test, bagging_predictions)
print(f"Bagging Classifier Accuracy: {bagging_accuracy}")
print(classification_report(y_test, bagging_predictions))

# Comparing the two models
if bagging_accuracy > dt_accuracy:
    print("The bagging model improved over the decision tree.")
else:
    print("The bagging model did not improve over the decision tree.")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris

# Load the iris dataset
iris_data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, test_size=0.3, random_state=42)

# Train the Decision Tree model
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(X_train, y_train)

# Train the Random Forest model
random_forest_model = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest_model.fit(X_train, y_train)

# Evaluate the Decision Tree model
dt_predictions = decision_tree_model.predict(X_test)
dt_accuracy = accuracy_score(y_test, dt_predictions)

# Evaluate the Random Forest model
rf_predictions = random_forest_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_predictions)

# Print the accuracy and classification report for both models
print("Decision Tree Model Accuracy:", dt_accuracy)
print("Random Forest Model Accuracy:", rf_accuracy)

# Detailed report
print("\nDecision Tree Classification Report:")
print(classification_report(y_test, dt_predictions))
print("\nRandom Forest Classification Report:")
print(classification_report(y_test, rf_predictions))

# Determine if there was an improvement
if rf_accuracy > dt_accuracy:
    print("The Random Forest model is an improvement over the Decision Tree model.")
else:
    print("The Random Forest model did not improve over the Decision Tree model.")
```

# **Integrating AdaBoost**
AdaBoost (Adaptive Boosting) works by combining multiple weak learning models to create a strong predictive model. Decision trees are commonly used as the base learner for AdaBoost.

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

def train_adaboost_model(iris):
    """
    Train an AdaBoost classifier using decision trees.

    :param iris: iris data set instance
    :return: trained AdaBoost classifier
    """
    base_clf = DecisionTreeClassifier()  # Base decision tree classifier
    ada_clf = AdaBoostClassifier(base_estimator=base_clf, n_estimators=50, learning_rate=1, random_state=0)
    ada_clf.fit(iris.data, iris.target)
    return ada_clf
```

```python
import pydotplus
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from IPython.display import Image, display

def load_data_set():
    """
    Loads the iris data set
    :return: data set instance
    """
    iris = load_iris()
    return iris

def train_decision_tree(iris):
    """
    Train decision tree classifier
    :param iris: iris data set instance
    :return: classifier instance
    """
    clf = DecisionTreeClassifier()
    clf = clf.fit(iris.data, iris.target)
    return clf

def train_adaboost_model(iris):
    """
    Train AdaBoost classifier
    :param iris: iris data set instance
    :return: trained AdaBoost classifier instance
    """
    base_estimator = DecisionTreeClassifier()
    # Update here: Use 'estimator' instead of 'base_estimator' if scikit-learn version is 1.2 or later
    ada_clf = AdaBoostClassifier(estimator=base_estimator, n_estimators=50, random_state=42)
    ada_clf.fit(iris.data, iris.target)
    return ada_clf

def display_image(clf, iris):
    """
    Displays the decision tree image
    :param clf: classifier instance
    :param iris: iris data set instance
    """
    dot_data = export_graphviz(clf, out_file=None, feature_names=iris.feature_names,
                               class_names=iris.target_names, filled=True, rounded=True)
    graph = pydotplus.graph_from_dot_data(dot_data)
    display(Image(data=graph.create_png()))

if __name__ == '__main__':
    iris_data = load_data_set()

    # Split the dataset into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, test_size=0.3, random_state=42)

    # Train the decision tree classifier
    decision_tree_classifier = train_decision_tree(iris_data)
    decision_tree_classifier.fit(X_train, y_train)

    # Train the AdaBoost classifier
    adaboost_classifier = train_adaboost_model(iris_data)
    adaboost_classifier.fit(X_train, y_train)

    # Evaluate the decision tree classifier
    dt_predictions = decision_tree_classifier.predict(X_test)
    dt_accuracy = accuracy_score(y_test, dt_predictions)
    print("Decision Tree Classifier Accuracy:", dt_accuracy)
    print(classification_report(y_test, dt_predictions))

    # Evaluate the AdaBoost classifier
    ab_predictions = adaboost_classifier.predict(X_test)
    ab_accuracy = accuracy_score(y_test, ab_predictions)
    print("AdaBoost Classifier Accuracy:", ab_accuracy)
    print(classification_report(y_test, ab_predictions))

    # Comparing performances
    if ab_accuracy > dt_accuracy:
        print("AdaBoost Model showed improvement over the Decision Tree Model.")
    else:
        print("AdaBoost Model did not show improvement over the Decision Tree Model.")
```

```python
import pydotplus
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from IPython.display import Image, display

def load_data_set():
    """
    Loads the iris data set
    :return: data set instance
    """
    iris = load_iris()
    return iris

def train_decision_tree(iris):
    """
    Train decision tree classifier
    :param iris: iris data set instance
    :return: classifier instance
    """
    clf = DecisionTreeClassifier()
    clf = clf.fit(iris.data, iris.target)
    return clf

def train_gradient_boosting_model(iris):
    """
    Train Gradient Boosting classifier
    :param iris: iris data set instance
    :return: trained Gradient Boosting classifier instance
    """
    gb_clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=42)
    gb_clf.fit(iris.data, iris.target)
    return gb_clf

def display_image(clf, iris):
    """
    Displays the decision tree image
    :param clf: classifier instance
    :param iris: iris data set instance
    """
    dot_data = export_graphviz(clf, out_file=None, feature_names=iris.feature_names,
                               class_names=iris.target_names, filled=True, rounded=True)
    graph = pydotplus.graph_from_dot_data(dot_data)
    display(Image(data=graph.create_png()))

if __name__ == '__main__':
    iris_data = load_data_set()

    # Split the dataset into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, test_size=0.3, random_state=42)

    # Train the decision tree classifier
    decision_tree_classifier = train_decision_tree(iris_data)
    decision_tree_classifier.fit(X_train, y_train)

    # Train the Gradient Boosting classifier
    gradient_boosting_classifier = train_gradient_boosting_model(iris_data)
    gradient_boosting_classifier.fit(X_train, y_train)

    # Evaluate the decision tree classifier
    dt_predictions = decision_tree_classifier.predict(X_test)
    dt_accuracy = accuracy_score(y_test, dt_predictions)
    print("Decision Tree Classifier Accuracy:", dt_accuracy)
    print(classification_report(y_test, dt_predictions))

    # Evaluate the Gradient Boosting classifier
    gb_predictions = gradient_boosting_classifier.predict(X_test)
    gb_accuracy = accuracy_score(y_test, gb_predictions)
    print("Gradient Boosting Classifier Accuracy:", gb_accuracy)
    print(classification_report(y_test, gb_predictions))

    # Comparing performances
    if gb_accuracy > dt_accuracy:
        print("The Gradient Boosting model showed improvement over the Decision Tree model.")
    else:
        print("The Gradient Boosting model did not show improvement over the Decision Tree model.")
```
