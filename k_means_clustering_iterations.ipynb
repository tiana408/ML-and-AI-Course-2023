Step 1: Assign each point to the nearest centroid. The distance can be calculated using the Euclidean distance formula.

```python
import numpy as np

# Data points
points = np.array([
    [1, 2],
    [1, 4],
    [1, 0],
    [4, 2],
    [4, 4],
    [4, 0]
])

# Initial centroids
centroids = np.array([
    [1, 2], # Centroid A
    [4, 2]  # Centroid B
])

# Function to calculate Euclidean distance
def euclidean_distance(p1, p2):
    return np.sqrt(np.sum((p1 - p2) ** 2))

# Assign points to the nearest centroid
assignments = []
for point in points:
    distances = [euclidean_distance(point, centroid) for centroid in centroids]
    closest_centroid = np.argmin(distances)
    assignments.append(closest_centroid)

# Output the assignments
for i, point in enumerate(points):
    print(f"Point {point} is assigned to centroid {assignments[i]+1}")
```

Iteration: Repeat steps 2 and 3 until the centroids no longer change significantly.

```python
import numpy as np

# Define the dataset points
points = np.array([
    [1, 2],
    [1, 4],
    [1, 0],
    [4, 2],
    [4, 4],
    [4, 0]
])

# Define the initial centroids
centroids = np.array([
    [1, 2],  # Centroid A
    [4, 2]   # Centroid B
])

# Function to calculate Euclidean distance
def euclidean_distance(point, centroid):
    return np.sqrt(np.sum((point - centroid) ** 2))

# Iterate until convergence
for _ in range(100):  # Limit the number of iterations to prevent infinite loops
    # Step 1: Assign each point to the nearest centroid
    assignments = []
    for point in points:
        distances = [euclidean_distance(point, centroid) for centroid in centroids]
        closest_centroid = np.argmin(distances)
        assignments.append(closest_centroid)

    # Step 2: Recalculate centroids
    new_centroids = np.array([np.mean(points[np.array(assignments) == i], axis=0) for i in range(2)])

    # Check if centroids have changed
    if np.all(centroids == new_centroids):
        break

    centroids = new_centroids

# Print the final centroids
print("Final Centroids:", centroids)
```
